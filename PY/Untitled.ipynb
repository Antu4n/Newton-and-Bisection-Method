```python
import math
import matplotlib.pyplot as plt
import timeit
import numpy as np
```

```python
def func (x):
     return x**4 - 2*x**3 - 5*x**2 + 4*x + 6
```

```python
a = 2
b = 3
tolerance = 1 * math.exp(-6)
max_iterations = 5000
```

```python
for i in range(max_iterations):
    c = (a + b) / 2
    if (abs(func(c)) < tolerance):
        print(f"Root found at x = {c:.6f}")
        print("The time taken is ",timeit.timeit())
        break
    elif (func(c) * func(a)) < 0:
        b = c
    else:
        a = c
```

```python
def deriv(x):
    return 15*x**4 + 8*x**3 + 10*x - 1
```

```python
x0 = 4
tolerance = 1e-6
max_iter = 100
```

```python
for i in range(max_iter):
    fx = func(x0)
    d_fx = deriv(x0)
    
    x1 = x0 -(fx/d_fx)
    if abs(func(x1))< tolerance:
        print(f"Root found at x = {x1:.6f}")
        print("The time taken taken is" ,timeit.timeit())
        break
    else:
        x0 = x1
```
